```python
import torch
from transformers import AutoModelForMaskedLM, AutoTokenizer
from huggingface_hub import login

login()
```

```python
dir = './ESMC-300M-CLI'
```

```python
sequences = ["MSKGEELFTGVVPILVELDGDVNGHKFSVSGEGEGDATYGKLTLKFICTTGKLPVPWPTLVTTFSYGVQCFSRYPDHMKQHDFFKSAMPEGYVQERTIFFKDDGNYKTRAEVKFEGDTLVNRIELKGIDFKEDGNILGHKLEYNYNSHNVYIMADKQKNGIKVNFKIRHNIEDGSVQLADHYQQNTPIGDGPVLLPDNHYLSTQSALSKDPNEKRDHMVLLEFVTAAGITHGMDELYK"]

model = AutoModelForMaskedLM.from_pretrained(
    dir, device_map="auto", local_files_only=True).eval()
```

```python
tokenizer = AutoTokenizer.from_pretrained(dir, local_files_only=True)

```

```python
inputs = tokenizer(sequences, return_tensors="pt", padding=True)
inputs = {k: v.to(model.device) for k, v in inputs.items()}
with torch.inference_mode():
    output = model(**inputs, output_hidden_states=True)
```

```python
logits = output.logits
print("Logits Shape:", logits.shape)
```

```python
embeddings = output.hidden_states[-1]
print("Embeddings Shape:", embeddings.shape)
```

```python
protein_representation = embeddings.mean(dim=1)
print("Protein Vector Shape:", protein_representation.shape)
```
